```python
'''
Q-18>What is the difference between os.fork() and multiprocessing in Python?
Sol:
os.fork() and the multiprocessing module are used for creating new processes .
os.fork():
A.workings:
1.It creates a child process by duplicating the current process (parent).
2.The child gets a copy of the parent's memory space.
B.Return Values:
1.Returns 0 in the child process.
2. Returns the child’s process ID (PID) in the parent process.
Main task:
1.Duplicates the parent process
2.Child inherits memory from parent
3.Only works on UNIX/Linux
4.Low-level, requires manual management
5.System-level process creation
Ex:
'''
import os
import sys
import multiprocessing
import multiprocessing
def worker():
    print(f"Worker Process: PID = {os.getpid()}")
if __name__ == "__main__":
    process = multiprocessing.Process(target=worker)
    process.start()
    process.join()
    print(f"Parent Process: PID = {os.getpid()}")

'''
1.multiprocessing:
A.Workings:
1.Uses the Process class to create new processes, instead of duplicating the current process.
2.The new process has its own memory space, so it doesn’t share variables with the parent.
Task:
1.Creates a new independent process
2.Each process has separate memory
3.Works on Windows, macOS, and Linux
4.High-level, easier to use
5.Parallel execution and task processing
Ex:
'''
import multiprocessing
def worker():
    print("Worker process: PID =", multiprocessing.current_process().pid)
if __name__ == "__main__":
    process = multiprocessing.Process(target=worker)
    process.start()
    process.join()
    print("Parent process: PID =", multiprocessing.current_process().pid)
```

```python
'''
Q-19>  Write a program that handles both IndexError and KeyError using a try-except block?
Sol:
'''
def handle_errors():
    my_list = [1,20,3]
    my_dict = {"Product": "RMV-098", "PRICE": 24530000}
    try:        
        print(f"List item: {my_list[2]}")        
        print(f"Country: {my_dict['country']}")
    except IndexError:
        print("Error: List index out of range.")
    except KeyError:
        print("Error: Dictionary key not found.")
handle_errors()
```

```python
'''
Q-20> How would you open a file and read its contents using a context manager in Python ?
Sol:
'''
def read_file(filename):
    try:
        with open(filename, "r") as file:
            content = file.read()
            print("File Contents:\n")
            print(content)
    except FileNotFoundError:
        print(f"Error: The file '{filename}' does not exist.")
    except Exception as e:
        print(f"An error occurred: {e}")
file_name = (r"C:\Users\hp\OneDrive\Desktop\sumit.txt")
read_file(file_name)
```

```python
'''
Q-21> Write a Python program that reads a file and prints the number of occurrences of a specific word ?
Sol:
'''
def count_word_occurrences(filename,target_word):
    try:
        with open(filename,"r") as file:
            content = file.read().lower()  
            words = content.split()  
            count = words.count(target_word.lower())  
            print(f"The word '{target_word}' appears {count} times in '{filename}'.")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' does not exist.")
    except Exception as e:
        print(f"An error occurred: {e}")
file_name=(r"C:\Users\hp\OneDrive\Desktop\sumit.txt")  
search="INFO"     
count_word_occurrences(file_name,search)
```

```python
'''
Q-23> Write a Python program that writes to a log file when an error occurs during file handling ?
Sol:
'''
import logging
logging.basicConfig(filename=r"C:\Users\hp\OneDrive\Desktop\sumit.txt",level=logging.ERROR,format="%(asctime)s - %(levelname)s - %(message)s")
def read_file(filename):
    try:
        with open(filename, "r") as file:
            Data = file.read()
            print("File Contents:\n", Data)
    except FileNotFoundError:
        error_message = f"Error: The file '{filename}' does not exist."
        print(error_message)
        logging.error(error_message)
    except Exception as e:
        error_message = f"An unexpected error occurred: {e}"
        print(error_message)
        logging.error(error_message)
file_name = (r"C:\Users\hp\OneDrive\Desktop\sumit.txt")  
read_file(file_name)
```
